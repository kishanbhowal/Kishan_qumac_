```python
import csv
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
V_init = -7.9
V_final = 7.9
V_step = 0.1
# V_arr = np.arange(V_init, V_final + V_ste, V_step)
V_arr = np.arange(V_init, V_final+V_step, V_step)
V_arr = np.round(V_arr, 1)
```

```python
V_arr
```

```python
data_2d = []
for V in V_arr:
    # filepath = rf'D:\Experiments\2023-11-17 2DRX08, Double Ring, New  Paramps Test\ParampsData\CPWPP06B\ParampPhaseResponse_{V}.txt'
    filepath = rf"D:\Experiments\2024-03-19 Ring with INDIQTa03, Cu with 3D08BC(withoutparamp)\ParampsData\Ring7\PhaseSweep2\ParampPhaseResponse_{V}.txt"
    print(filepath)
    data= np.transpose(np.loadtxt(filepath))
    ydata=data[1]
    data_2d.append(ydata)
freq = data[0]
```

```python
V_ticks = np.round(np.arange(V_init, V_final-0.5, -1), 2)
V_ticks
```

```python
ax = plt.figure(figsize=(15, 10))
cmap = ['RdBu_r', 'YlGnBu', 'BuPu']
sns.set(font_scale=1)
ax = sns.heatmap(data_2d, annot=False,
                 xticklabels=np.round(1e-9*freq, 3), yticklabels=V_arr,
                 cmap=cmap[0], cbar_kws={'label': 'Degrees'})
ax.set_xlabel("Frequency (GHz)")
ax.set_ylabel("Voltage (V)")
ax.set_title("Paramp Phase Response - CPWPP06B")

for index, label in enumerate(ax.get_yticklabels()):
   if index % 10 == 0:
      label.set_visible(True)
   else:
      label.set_visible(False)

for index, label in enumerate(ax.get_xticklabels()):
   if index % 50 == 0:
      label.set_visible(True)
   else:
      label.set_visible(False)
plt.savefig(r"D:\Experiments\2024-03-19 Ring with INDIQTa03, Cu with 3D08BC(withoutparamp)\ParampsData\Ring7\PhaseResponse2_heatmap.pdf")
```
